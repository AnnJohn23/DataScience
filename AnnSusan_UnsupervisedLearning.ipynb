```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage,dendrogram

from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline #pipeline means sequence of action

from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score 
```

```python

#read the dataset
df = pd.read_csv('Wine_clust.csv')
df.head()
```

# Try out different clustering models

# 1.K-Means Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42) # no. of clusters taken 3 as random
kmeans.fit(df)
label = kmeans.predict(df)
label
```

```python
# scatter plot of any 2 features.
Alcohol = df['Alcohol']
Ash = df['Ash']
plt.scatter(Alcohol,Ash, c=label)
plt.show()
```

```python
kmeans_score = silhouette_score(df, kmeans.labels_, metric='euclidean')
print('Silhouette Score: %.3f' % kmeans_score)
```

# 2.Agglomerative Hierarchical Clustering (AHC)

```python
ahc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
labels = ahc.fit_predict(df)
labels
```

```python
# scatter plot of any 2 features.
Alcohol = df['Alcohol']
Ash = df['Ash']
plt.scatter(Alcohol,Ash, c=labels)
plt.show()
```

```python
ahc_score = silhouette_score(df, ahc.labels_, metric='euclidean')
print('Silhouette Score: %.4f' % ahc_score)
```

```python
#Findings:
1)We can see that they are some variations in clustering between two clustering method
2)K-means model have a silhoette score of 0.571 and that of AHC model is 0.5645
```

# The optimum number of clusters in each model

# 1.For finding optimum no. of clusters we can do elbow method for K-means clustering.

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,'-o')
plt.title('Elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS Value')
plt.show()
```

# Findings:

We can take optimum value of k as 3, as per the elbow method.

# 2. For finding optimum no. of clusters in AHC, we can make use of dendrogram.

```python
link = linkage(df,method = 'ward') 
plt.figure(figsize=(15,8))
dendrogram(link,orientation='top',distance_sort='ascending',show_leaf_counts=True)
plt.hlines(3500,0,5000,colors="red",linestyles="dashed")
plt.show()
```

# Findings:

While observing dendrogram, we can figure out that optimum no. of cluster is 2 for AHC model.

```python
# builidng model with no. of clusters as 2
ahc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
labels = ahc.fit_predict(df)
labels
```

```python
new_ahc_score = silhouette_score(df, ahc.labels_, metric='euclidean')
print('Silhouette Score: %.4f' % new_ahc_score)
```
